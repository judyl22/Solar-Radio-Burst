```python
bursts_cleaned = bursts[bursts["Brightening"].notna()]

duration_data = bursts_cleaned["Duration (seconds)"]
drift_rate_data = bursts_cleaned["Drift Rate (Hz/s)"]
brightening_data = bursts_cleaned["Brightening"].unique()

log_bins_duration = np.logspace(np.log10(duration_data.min()), np.log10(duration_data.max()), 20)
log_bins_drift_rate = np.logspace(np.log10(drift_rate_data.min()), np.log10(drift_rate_data.max()), 20)

colors = ['blue', 'orange', 'green']

fig, ax = plt.subplots(1, 2, figsize=(15, 5))

for i, brightening_type in enumerate(brightening_data):
    subset_duration = duration_data[bursts_cleaned["Brightening"] == brightening_type]
    hist_duration, bin_edges_duration = np.histogram(subset_duration, bins=log_bins_duration)
    
    ax[0].bar(bin_edges_duration[:-1], hist_duration, width=np.diff(bin_edges_duration), 
              alpha=0.5, color=colors[i], label=brightening_type, edgecolor="black")

ax[0].set_xscale('log')
ax[0].set_title('Log-Binned Histogram for Duration (seconds)')
ax[0].set_xlabel('Duration (seconds) (log scale)')
ax[0].set_ylabel('Frequency')
ax[0].legend(title='Brightening')

for i, brightening_type in enumerate(brightening_data):
    subset_drift_rate = drift_rate_data[bursts_cleaned["Brightening"] == brightening_type]
    hist_drift_rate, bin_edges_drift_rate = np.histogram(subset_drift_rate, bins=log_bins_drift_rate)
    
    ax[1].bar(bin_edges_drift_rate[:-1], hist_drift_rate, width=np.diff(bin_edges_drift_rate), 
              alpha=0.5, color=colors[i], label=brightening_type, edgecolor="black")

ax[1].set_xscale('log')
ax[1].set_title('Log-Binned Histogram for Drift Rate (Hz/s)')
ax[1].set_xlabel('Drift Rate (Hz/s) (log scale)')
ax[1].set_ylabel('Frequency')
ax[1].legend(title='Brightening')

plt.tight_layout()
plt.show()
```
