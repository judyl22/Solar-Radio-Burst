```python
import numpy as np
import pandas as pd

file_path = 'stereo_maven_psp_both.csv'

bursts = pd.read_csv(file_path)

print(bursts.head())
```

```python
stereo = bursts.iloc[:, 0:5]
stereo
```

```python
psp = bursts.iloc[:, 5:10]
psp.columns = [col.replace('.1', '') for col in psp.columns]
psp
```

```python
def calc_drift_rate(bursts):
    freq_change = bursts["High frequency (Hz)"] - bursts["Low frequency (Hz)"]
    bursts["Frequency Change"] = freq_change
    
    bursts['Start time'] = pd.to_datetime(bursts['Start time'], format='%Y-%m-%dT%H:%M', errors='coerce')
    bursts['End time'] = pd.to_datetime(bursts['End time'], format='%Y-%m-%dT%H:%M', errors='coerce')
    bursts['Duration (seconds)'] = (bursts['End time'] - bursts['Start time']).dt.total_seconds()
    
    drift_rate = bursts["Frequency Change"] / bursts["Duration (seconds)"]
    bursts["Drift Rate (Hz/s)"] = drift_rate
calc_drift_rate(stereo)
calc_drift_rate(psp)
stereo.head()
```

```python
# generate pop-up plot
from pytplot import get_data
from matplotlib import pyplot as plt
%matplotlib tk
from matplotlib.colors import LogNorm
```

```python
# function to plot duration histograms with error bars
def plot_duration_hist(data1, name1, data2, name2):
    # Extract and clean duration data
    duration1 = data1["Duration (seconds)"].dropna()
    duration2 = data2["Duration (seconds)"].dropna()

    # Determine the combined range for both datasets
    combined_min = min(duration1.min(), duration2.min())
    combined_max = max(duration1.max(), duration2.max())

    # Define logarithmic bins based on the combined range
    log_bins = np.logspace(np.log10(combined_min), np.log10(combined_max), 20)

    plt.figure(figsize=(10, 6))

    # Plot histograms for both datasets
    counts1, bins1, patches1 = plt.hist(duration1, bins=log_bins, edgecolor='black', alpha=0.5, label=name1 + ' Histogram', color='blue')
    counts2, bins2, patches2 = plt.hist(duration2, bins=log_bins, edgecolor='black', alpha=0.5, label=name2 + ' Histogram', color='green')

    # Calculate bin centers
    bin_centers = (log_bins[:-1] + log_bins[1:]) / 2

    # Plot line plots of counts
    plt.plot(bin_centers, counts1, label=name1 + ' Counts', color='blue')
    plt.plot(bin_centers, counts2, label=name2 + ' Counts', color='green')

    # Plot error bars
    plt.errorbar(bin_centers, counts1, yerr=np.sqrt(counts1), fmt='o', color='blue', alpha=0.3)
    plt.errorbar(bin_centers, counts2, yerr=np.sqrt(counts2), fmt='o', color='green', alpha=0.3)

    # Set logarithmic scale for x-axis
    plt.xscale('log')
    plt.xlim([combined_min * 0.8, combined_max * 1.2])  # Add padding
    plt.title('Overlapping Duration Histograms with Error Bars')
    plt.xlabel('Duration (s)')
    plt.ylabel('Count')
    plt.grid(True, which="both", ls="--", linewidth=0.5)
    plt.legend()
    plt.show()

# Example usage with 'stereo' and 'psp' DataFrames
plot_duration_hist(stereo, "Stereo", psp, "Parker Solar Probe")
```

```python
stereo["Start Date"] = pd.to_datetime(stereo["Start time"]).dt.date
psp["Start Date"] = pd.to_datetime(psp["Start time"]).dt.date

plt.figure(figsize=(10, 6))

# Scatter plot for stereo data (blue)
plt.scatter(stereo["Duration (seconds)"], stereo["Frequency Change"], alpha=0.7, label="Stereo", color="blue")

# Scatter plot for psp data (red)
plt.scatter(psp["Duration (seconds)"], psp["Frequency Change"], alpha=0.7, label="PSP", color="red")

# Annotate points with Start Date
for i, txt in enumerate(stereo["Start Date"]):
    plt.annotate(txt, (stereo["Duration (seconds)"].iloc[i], stereo["Frequency Change"].iloc[i]), fontsize=8, alpha=0.75)

for i, txt in enumerate(psp["Start Date"]):
    plt.annotate(txt, (psp["Duration (seconds)"].iloc[i], psp["Frequency Change"].iloc[i]), fontsize=8, alpha=0.75)
    
# Set logarithmic scale (base 10) for both axes
plt.xscale("log")
plt.yscale("log")

plt.xlabel("Duration (seconds)")
plt.ylabel("Frequency Change (Hz)")
plt.title("Frequency Change vs Duration (Stereo & PSP)")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

```python
# updated version with adjust

# Extract Start Date (already done in your code)
stereo["Start Date"] = pd.to_datetime(stereo["Start time"]).dt.date
psp["Start Date"] = pd.to_datetime(psp["Start time"]).dt.date

# Begin plotting
plt.figure(figsize=(10, 6))
plt.scatter(stereo["Frequency Change"], stereo["Duration (seconds)"], alpha=0.7, label="Stereo", color="blue")
plt.scatter(psp["Frequency Change"], psp["Duration (seconds)"], alpha=0.7, label="PSP", color="red")

# Collect labels to adjust
texts = []
for i, txt in enumerate(stereo["Start Date"]):
    texts.append(plt.text(stereo["Frequency Change"].iloc[i], stereo["Duration (seconds)"].iloc[i], str(txt),
                          fontsize=8, alpha=0.75, color='blue'))

for i, txt in enumerate(psp["Start Date"]):
    texts.append(plt.text(psp["Frequency Change"].iloc[i], psp["Duration (seconds)"].iloc[i], str(txt),
                          fontsize=8, alpha=0.75, color='red'))

# Adjust text to prevent overlap
adjust_text(texts, arrowprops=dict(arrowstyle='-', color='gray', lw=0.5))

# Set log scale
plt.xscale("log")
plt.yscale("log")

# Labels, title, legend
plt.xlabel("Frequency Change (Hz) (log scale)")
plt.ylabel("Duration (seconds) (log scale)")
plt.title("Log–Log Scatter Plot of Duration vs Frequency Change (Stereo & PSP)")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.tight_layout()
plt.show()
```
