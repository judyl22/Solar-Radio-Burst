```python
import pandas as pd
```

```python
file_path = 'stereo_maven.csv'
bursts = pd.read_csv(file_path)
print(bursts.head())
```

```python
bursts["Start time"] = pd.to_datetime(bursts["Start time"])
bursts["End time"] = pd.to_datetime(bursts["End time"])
bursts["Duration (s)"] = (bursts["End time"] - bursts["Start time"]).dt.total_seconds()
bursts["Frequency Change"] = bursts["High frequency (Hz)"] - bursts["Low frequency (Hz)"]
bursts["Drift Rate (Hz/s)"] = bursts["Frequency Change"] / bursts["Duration (s)"]
```
